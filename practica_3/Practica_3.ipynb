```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

## Creación de los nodos

```python
def generar_red_aleatoria(N, P):
    """
    Genera una red aleatoria no dirigida con N nodos y probabilidad P de enlace entre nodos.

    Parámetros:
    N (int): Número de nodos de la red.
    P (float): Probabilidad de que exista un enlace entre cualquier par de nodos.

    Retorna:
    G (networkx.Graph): Un objeto Graph con la red aleatoria no dirigida.
    """
    G = nx.Graph()
    G.add_nodes_from(range(N))
    for i in range(N):
        for j in range(i+1, N):  # Evitar bucles y enlaces múltiples
            if random.random() < P:
                G.add_edge(i, j)
    return G
```
